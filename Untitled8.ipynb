```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

# Upload the file
uploaded = files.upload()

# Read the uploaded Excel file
df = pd.read_csv(list(uploaded.keys())[0])
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])  # Convert to datetime
#df.sort_values('Date', inplace=True)
plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['Total Revenue'], marker='o', linestyle='-', color='b')
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.title("Sales Trend Over Time")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
category_sales = df.groupby("Product Category")["Total Revenue"].sum().sort_values()

plt.figure(figsize=(8,5))
plt.bar(category_sales.index, category_sales.values, color='c')
plt.xlabel("Revenue")
plt.ylabel("Category")
plt.title("Total Revenue by Product Category")
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Total Revenue'], bins=20, color='green', alpha=0.7, edgecolor='black')
plt.xlabel("Total Revenue")
plt.ylabel("Frequency")
plt.title("Revenue Distribution")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
region_sales = df.groupby("Region")["Total Revenue"].sum()

plt.figure(figsize=(7,7))
plt.pie(region_sales, labels=region_sales.index, autopct='%1.1f%%', colors=['gold', 'lightcoral', 'lightskyblue'])
plt.title("Sales Distribution by Region")
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(8, 5))
sns.histplot(df['Total Revenue'], kde=True, bins=20)
plt.title("Distribution of Total Revenue")
plt.xlabel("Total Revenue")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Product Category', y='Total Revenue', data=df)
plt.xticks(rotation=45)
plt.title("Total Revenue by Product Category")
plt.show()
```

```python
from numpy import mean

plt.figure(figsize=(8, 5))
sns.barplot(x='Region', y='Total Revenue', data=df, estimator=mean)
plt.title("Average Total Revenue by Region")
plt.ylabel("Average Revenue")
plt.show()
```

```python
daily_revenue = df.groupby('Date')['Total Revenue'].sum().reset_index()

plt.figure(figsize=(7, 3))
sns.lineplot(data=daily_revenue, x='Date', y='Total Revenue')
plt.title("Daily Total Revenue Over Time")
plt.ylabel("Total Revenue")
plt.xlabel("Date")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Payment Method', data=df)
plt.title("Transaction Count by Payment Method")
plt.ylabel("Count")
plt.show()
```

```python
# top 10 products based on total revenue
top_products = df.groupby('Product Name')['Total Revenue'].sum().nlargest(10).reset_index()

plt.figure(figsize=(7, 3))
sns.barplot(data=top_products, x='Total Revenue', y='Product Name')
plt.title("Top 10 Products by Revenue")
plt.xlabel("Total Revenue")
plt.ylabel("Product Name")
plt.show()
```

```python
#violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='Region', y='Total Revenue', data=df)
plt.title("Revenue Distribution by Region")
plt.show()
```
