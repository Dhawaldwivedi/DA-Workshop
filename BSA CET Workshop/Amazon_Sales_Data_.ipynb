# **Data Loading and Cleaning**

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
````

````python
df = pd.read_csv('/content/Amazon Sales.csv')
````

````python
# taking the top 5 values
df.head()
````

````python
df.shape
````

````python
df.info()
````

````python
df.describe()
````

**Checking for the NULL values**

````python
df.isnull().sum()
````

````python
# changing the date type
df['order_date'] = pd.to_datetime(df['order_date'])
````

````python
df.info()
````

````python
# creating some helpful columns
df["month"] = df["order_date"].dt.month
df["year"] = df["order_date"].dt.year
````

````python
df.head()
````

# **Analysis**

````python
# Price Distribution
plt.figure(figsize=(8,5))
sns.histplot(df["price"], bins=30)
plt.title("Price Distribution")
plt.show()
````

````python
# Discount Distribution
plt.figure(figsize=(8,5))
sns.histplot(df["discount_percent"], bins=20)
plt.title("Discount Percentage Distribution")
plt.show()
````

````python
# Category Analysis
category_counts = df["product_category"].value_counts()

plt.figure(figsize=(8,5))
category_counts.plot(kind="bar")
plt.title("Orders by Product Category")
plt.xlabel("Category")
plt.ylabel("Number of Orders")
plt.show()
````

````python
revenue_category = df.groupby("product_category")["total_revenue"].sum()

plt.figure(figsize=(8,5))
revenue_category.sort_values().plot(kind="barh")
plt.title("Revenue by Product Category")
plt.show()
````

````python
region_sales = df.groupby("customer_region")["total_revenue"].sum()

plt.figure(figsize=(7,7))
region_sales.plot(kind="pie", autopct='%1.1f%%')
plt.title("Revenue Share by Region")
plt.ylabel("")
plt.show()
````

````python
payment_sales = df.groupby("payment_method")["total_revenue"].sum()

plt.figure(figsize=(8,5))
payment_sales.sort_values().plot(kind="bar")
plt.title("Revenue by Payment Method")
plt.show()
````

````python
monthly_sales = df.groupby(df["order_date"].dt.to_period("M"))["total_revenue"].sum()

monthly_sales.plot(figsize=(10,5))
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
````

#**Performing Analysis using AI**

````python
# Installing necessary Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import google.generativeai as genai
from google.colab import userdata
````

**Loading the API key ( through colab secrets )**

````python

# -----------------------------
api_key = userdata.get("GOOGLE_API_KEY")
genai.configure(api_key=api_key)
````

**Initialize Gemini model**

````python

model = genai.GenerativeModel("gemini-2.5-flash")
````

**Loading Dataset and Generating Basic Summary**

````python

# -----------------------------
df = pd.read_csv("Amazon Sales.csv")

print("DATASET SHAPE:", df.shape)
print("\nCOLUMNS:", list(df.columns))

# -----------------------------
print("\nMISSING VALUES\n", df.isnull().sum())

print("\nSTATISTICAL SUMMARY\n")
print(df.describe())

````

**Prepare Dataset Context**

````python

# -----------------------------
dataset_info = f"""
Dataset Shape: {df.shape}

Columns: {list(df.columns)}

Data Types:
{df.dtypes}

Missing Values:
{df.isnull().sum()}

Sample Data:
{df.head(10).to_string()}
"""
````

**Ask AI for Insights**

````python

# -----------------------------
analysis_prompt = f"""
You are a senior data scientist.

Analyze this dataset and provide:

1. Key patterns
2. Important correlations
3. Possible anomalies
4. Business insights
5. 5 analytical questions from the dataset

For each question also give:
- Python pandas code
- Expected answer/interpretation
"""

analysis_response = model.generate_content(analysis_prompt + dataset_info)

print("\n AI DATA ANALYSIS REPORT\n")
print(analysis_response.text)
````

**Generating Plots through AI**




````python
# -----------------------------
# 6. Ask AI to Generate EDA Plots
# -----------------------------
plot_prompt = f"""
You are a Python data analyst.

Dataset columns:
{list(df.columns)}

Write Python code using pandas, matplotlib, and seaborn to create 4 useful EDA visualizations.

Rules:
- Use the dataframe name 'df'
- Only return executable Python code
- Do not generate sample data
- Do not redefine df
- Do not include markdown or explanations
"""

plot_response = model.generate_content(plot_prompt)

ai_code = plot_response.text

# -----------------------------
# CLEAN THE AI CODE
# -----------------------------
ai_code = ai_code.replace("```python", "").replace("```", "").strip()

print("\nAI GENERATED VISUALIZATION CODE\n")
print(ai_code)

# -----------------------------
# 7. Execute AI Generated Code
# -----------------------------
print("\nRUNNING AI GENERATED VISUALIZATIONS\n")

try:
    exec(ai_code)
except Exception as e:
    print("Error running AI code:", e)
````
